### Get started

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints, color = 'green');
```

### plotting without line

```python
plt.plot(xpoints, ypoints, 'o');
```

### multiple point

```python
x = np.array([1,7,6,8,1])
y = np.array([8,2,7,1,3])

plt.plot(x,y);
```

### Default x points

```python
plt.plot(y);
```

### Make each point with a circle

```python
plt.plot(y, marker = 'o');
```

### make each point with star


```python
plt.plot(y, marker='*');
```

### Format strings fmt

```python
plt.plot(x, y, 'o:r'); #(Marker|line|color)      (o, : , red)
```

### Set the size of the maker

```python
plt.plot(y, marker='*', ms = 20);
```

### Maker color

```python
plt.plot(x, y, marker="*", ms=20, mec = 'r');
```

```python
plt.plot(x, y, marker = 'o', ms=20, mfc = 'red');
```

```python
plt.plot(x, y , marker='o', ms = 20, mec = 'red', mfc = 'green');
```

### Linestyle

```python
plt.plot(x, y, linestyle='--');
```

```python
plt.plot(x, y, linestyle = '--', color = 'red');
```

```python
plt.plot(x, y, linewidth = '10');
```

### multiple line

```python
plt.plot(x, y , color = 'red', linewidth = '10');
plt.plot(y, color= 'purple', linestyle = '--');
```

```python
plt.plot(x, y , y);
```

### set label

```python
plt.plot(x, y)
plt.title('test')
plt.xlabel('truc x')
plt.ylabel('truc y');
```

```python
plt.plot(x, y)
plt.title('test', loc='left')
plt.xlabel('truc x')
plt.ylabel('truc y');
```

### set grid

```python
plt.plot(x, y)
plt.grid()
```

```python
plt.plot(x, y)
plt.grid(axis='x')
```

```python
plt.plot(x, y)
plt.grid(color = 'purple',  linestyle = '--')
```

### draw multiple plots

```python
plt.subplot(1, 2, 1)
plt.plot(y)
plt.subplot(1, 2, 2)
plt.plot(x, y, color = 'red');
```

```python
x = np.random.randint(1,10, size=20)
y = np.random.randint(1,10, size=20)

plt.scatter(x, y);
```

```python
plt.scatter(x, y);
plt.scatter(y, x);
```

```python
colors = np.random.randint(1,100, size=20)
plt.scatter(x, y, c =  colors, cmap='viridis');
```

```python
colors = np.random.randint(1,100, size=20)
plt.scatter(x, y, c =  colors, cmap='viridis');
plt.colorbar()
```

```python
x = np.random.randint(1,100, size=100)
y = np.random.randint(1,100, size=100)
colors = np.random.randint(1,100, size=100)
sizes = np.random.randint(1,1000, size=100)
plt.scatter(x, y, c =  colors, s = sizes, cmap='viridis', alpha=0.3);
plt.colorbar();
```

### Bars

```python
x = np.array(['linh', 'nam', 'tuan'])
y = np.array([19,20,21])

plt.bar(x, y, width=0.2);
plt.xlabel('name')
plt.ylabel('age')
plt.title('Class');
```

```python
names = np.array(['banana', 'apple', 'orange'])
prices = np.array([20, 8 , 9])

plt.barh(names, prices, color= 'red');
```

### Histogram

```python
x = np.random.normal(100,20,100)
plt.hist(x);
```

### Pie charts

```python
a = np.array([80,10,60,20])
label = np.array(['banana', 'apple', 'orange', 'mango'])
plt.pie(a, labels=label);
```

```python
my_explode = np.array([0.05,0.05,0,0])
plt.pie(a, labels=label, explode=my_explode);
plt.legend(title = 'fruit');
```
