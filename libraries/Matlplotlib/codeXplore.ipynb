# Pyplot API and Object-oriented API
* Pyplot API -> Quickly
* OO API -> Advanced

### Pyplo API

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "sin(x)");
plt.plot(x, np.cos(x), color= "red", label ="cos(x)")

plt.title("Sin(X) and Cos(X)")
plt.xlabel("Bien x")
plt.ylabel("F(x)")
# plt.xlim([0,4]) # cut the chart from 0 - 4
# plt.ylim([-0.75,0.75])
# plt.axis([xmin, xmax, ymin, ymax]) # like xlim and ylim
# plt.axis("equal")
plt.legend(); # Display label ( cos and sin)
```

### Object-oriented API

```python
#1. Prepare data
x = [ 1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig ,ax = plt.subplots(figsize = (5,5)) # Figure size = Width and Height of plot

#3. Plot the data
ax.plot(x, y);
ax.set(title="A simple plot",
       xlabel = "x-axis",
       ylabel ="y-axis");
```

```python
# OO API

# prepare new data
rd = np.random.RandomState(0)
x = rd.randn(100)
y = rd.randn(100)

colors = rd.randn(100)
size = 1000* rd.randn(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s= size, c = colors, cmap='viridis', alpha = 0.5); # alpha: transparent
fig.colorbar(img1); # create color bar
```

### 3. Bar
* vertical
* horizontal


```python
# Prepare data from python dictionary
sort_drink_prices = {"Coke": 10,
                     "Pepsi": 12,
                     "Sprite": 8}
fig, ax = plt.subplots()
ax.bar(sort_drink_prices.keys(), sort_drink_prices.values());
ax.set(title="Bach hoa xanh", ylabel="Price ($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(sort_drink_prices.keys()), list(sort_drink_prices.values()))
```

### Subplots 

```python
# Option 1: 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                       ncols=2,
                       figsize= (10,5))
# Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(sort_drink_prices.keys(), sort_drink_prices.values());
ax4.hist(student_height, bins=30); # bins: chia nho cot x
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot data to each axis
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(sort_drink_prices.keys(), sort_drink_prices.values());
ax[1,1].hist(student_height, bins=30); # bins: chia nho cot x
```

# Project: Data Visualization - California cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./data/california_cities.csv')
cities.head()
```

```python
# Extract Latd ( Vi do) and Longd (Kinh do)

lat, lon = cities['latd'], cities['longd']

population, area = cities['population_total'], cities['area_total_km2']

plt.style.use('seaborn');
plt.figure(figsize=(8,6));
# Plot using pyplot API
plt.scatter(lon, lat,           
            c = np.log10(population), cmap='viridis',  # cmap lam mau me
            s = area, linewidths=0, alpha=0.5);       

plt.axis('equal');
plt.xlabel('longitude');
plt.ylabel('latitude');
plt.colorbar(label = 'log$_{10}$(population)');
plt.clim(3,7)       # limit color bar

# Create a legend for cities'sizes

area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([], [], s = area, c= 'k', alpha=0.4,
                label = str(area) + 'km$^2$')
plt.legend(labelspacing =1, title = 'City area')  # labelspacing cach nhau label, 

plt.title(' California Cities: Population and Area Distribution')
```
